```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../data/processed/players_final.csv")

df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
plt.figure(figsize=(12,6))

df["team"].value_counts().plot(
    kind="bar"
)

plt.title("Number of Players by Country")
plt.xlabel("Country")
plt.ylabel("Players")

plt.show()
```

```python
df["position"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Player Position Distribution")

plt.show()
```
